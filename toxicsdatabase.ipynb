```python
import os
import pandas as pd

# List all files in the folder
all_files = os.listdir(csv_folder)

# Filter for CSV files only
csv_files = [f for f in all_files if f.endswith(".csv")]

# Columns you want to extract from each CSV file
columns_to_extract = ["SAMPLENAME", "SAMPDATE", "ANALYTE", "RESULT", "UNITS"]

# Initialize an empty list to store DataFrames
dataframes = []

# Read specific columns from each CSV file and store in the list
for csv_file in csv_files:
    csv_path = os.path.join(csv_folder, csv_file)

    # Check if the CSV file has the expected columns
    try:
        # Read the CSV file and handle header row
        df = pd.read_csv(csv_path, header=0)
        
        # Check if all columns_to_extract are present in the DataFrame
        if all(col in df.columns for col in columns_to_extract):
            # Reorder columns if necessary
            df = df[columns_to_extract]
            
            dataframes.append(df)
        else:
            print(f"Skipping {csv_file} due to missing or unexpected columns.")
    except pd.errors.EmptyDataError:
        print(f"Skipping {csv_file} because it does not contain data.")
    except pd.errors.ParserError:
        print(f"Skipping {csv_file} because it has formatting issues or unexpected columns.")

    #df = pd.read_csv(csv_path, usecols=columns_to_extract)
    #dataframes.append(df)
```

```python
# Concatenate all DataFrames into a single DataFrame
combined_df = pd.concat(dataframes, ignore_index=True)
```
